```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_rating_prediction=pd.read_csv("/content/IMDb Movies India.csv",encoding='ISO-8859-1')
```

```python
movie_rating_prediction=movie_rating_prediction.dropna()
```

```python
movie_rating_prediction_encoded = pd.get_dummies(movie_rating_prediction, drop_first=True)
```

```python
movie_rating_prediction
```

```python
x = movie_rating_prediction_encoded.drop("Rating", axis=1)
y = movie_rating_prediction_encoded["Rating"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(train_x,train_y)
```

```python
predictions = model.predict(test_x)
print("Mean Squared Error:", mean_squared_error(test_y, predictions))
print("R2 Score:", r2_score(test_y, predictions))
```

```python
plt.bar(test_y, predictions, color='yellow',edgecolor="red")
plt.xlabel("Actual Rating")
plt.ylabel("Predicted Rating")
plt.title("Movie Ratings")
plt.show()
```
